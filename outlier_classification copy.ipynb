```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.preprocessing import PowerTransformer,StandardScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_excel('Dataset_GMSalary_outliers.xlsx')
df.head()
```

```python
df_2=df.iloc[:,1:]
df_2.describe()
```

```python
df_2=df_2.drop(112)
```

```python
df_2.describe()
```

```python
df_2['EDegree'] = df_2['EDegree'].apply(lambda x: 1 if x>1 else 0)
df_2['PTotal']=df_2[['PMotor','PHousehold','PHealth','PLife','PCompensations']].sum(axis=1)
```

```python
df_2['R_Motor'] = df_2['PMotor']/df_2['PTotal']
df_2['R_Household'] = df_2['PHousehold']/df_2['PTotal']
df_2['R_Health'] = df_2['PHealth']/df_2['PTotal']
df_2['R_Life'] = df_2['PLife']/df_2['PTotal']
df_2['R_Work_compensate'] = df_2['PCompensations']/df_2['PTotal']
```

```python
df_train=pd.read_excel('kprototype_cluster.xlsx')
```

```python
X=df_train.iloc[:,:-1].values
y=df_train.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
X_out=df_2.iloc[:,:].values
X_out
X_out = sc.fit_transform(X_out)
```

```python
y_outlier_pred = classifier.predict(X_out)
y_outlier_pred

```

```python
df_2['labels']=y_outlier_pred
```
